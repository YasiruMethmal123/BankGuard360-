```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"],dayfirst = True)
```

```python
df["Age"] = 2025 - df["Year_Birth"]
```

```python
spend_cols = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntSweetProducts","MntGoldProds"]
```

```python
df["Total_Spending"] = df[spend_cols].sum(axis = 1)
```

```python
df[["Total_Spending"]]
```

```python
df["Customer_Since"] = (pd.Timestamp("today") - df["Dt_Customer"]).dt.days
```

```python
corr = df[["Income","Age","Recency","Total_Spending","NumWebPurchases","NumStorePurchases"]]
corr
```

```python
pivot_income = df.pivot_table(values = "Income" , index = "Education" , columns = "Marital_Status" , aggfunc = "mean")
pivot_income
```

```python
sns.heatmap(pivot_income, annot = True , fmt = ".0f",cmap = "magma" )
plt.title("Avg INcome")
plt.show()
```

```python
group1 = df.groupby("Education")["Total_Spending"].mean().sort_values(ascending = False)
group1
```

```python
group1.plot(kind = "bar",color = "skyblue")
plt.title("Average spending by Education")
plt.ylabel("Average Total Spending")
plt.xticks(rotation = 45)
plt.show()
```

```python
df.columns
```

```python
df["AcceptedAny"] = df[["AcceptedCmp1","AcceptedCmp2","AcceptedCmp3","AcceptedCmp4","AcceptedCmp5"]].sum(axis = 1)
df["AcceptedAny"].unique
```

```python
group2 = df.groupby("Marital_Status")["AcceptedAny"].mean().sort_values(ascending = False)
group2
```

```python
group2.plot(kind = "bar" , color = "orange")
plt.title("Campain Acceptance rate by Martial Status")
plt.ylabel("Acceptence Rate")
plt.xticks(rotation = 45)
plt.show
```

```python
bins = [10,30,40,50,60,70,80]
labels = ["18-29","30-39","40-49","50-59","60-69","70+"]
df["AgeGroup"] = pd.cut(df["Age"],bins = bins,labels = labels)
df["AgeGroup"]
```

```python
group3 = df.groupby("AgeGroup")["Income"].mean()
```

```python
group3
```

```python
group3.plot(kind = "barh" , color = "green")
plt.title("Average Income by Age Group")
plt.xlabel("Average Income")
plt.show()
```
