```python
df["Dt_Customer"] = pd.to_datetime(df["Dt_Customer"],dayfirst = True)
```

```python
df.info()
```

```python
df["Age"] = 2025 - df["Year_Birth"]
```

```python
df["Age"]
```

```python
df.head()
```

```python
df["Total_Children"] = df["Kidhome"] + df["Teenhome"]
```

```python
df["Total_Children"]
```

```python
df.columns
```

```python
spend_cols = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts","MntSweetProducts","MntSweetProducts","MntGoldProds"]
```

```python
df["Total_Spending"] = df[spend_cols].sum(axis = 1)
```

```python
df[["Total_Spending"]]
```

```python
df["Customer_Since"] = (pd.Timestamp("today") - df["Dt_Customer"]).dt.days
```

```python
df["Customer_Since"]
```

```python
sns.histplot(df["Age"],bins = 30,kde = True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df["Income"],bins = 30,kde = True)
plt.title("Income Distribution")
plt.show()
```

```python
sns.histplot(df["Total_Spending"],bins = 30,kde = True)
plt.title("Total Spending Distribution")
plt.show()
```

```python
Catagorical Analysis
```

```python
sns.boxplot(x = "Education",y = "Income",data = df)
plt.xticks(rotation = 45)
plt.title("Income by Education")
plt.show()
```

```python
sns.boxplot(x = "Marital_Status",y = "Total_Spending",data = df)
plt.xticks(rotation = 45)
plt.title("Total Spending by Martal spending")
plt.show()
```

```python
corr = df[["Income","Age","Recency","Total_Spending","NumWebPurchases","NumStorePurchases"]]
corr
```

```python
pivot_income = df.pivot_table(values = "Income" , index = "Education" , columns = "Marital_Status" , aggfunc = "mean")
pivot_income
```

```python
sns.heatmap(pivot_income, annot = True , fmt = ".0f",cmap = "magma" )
plt.title("Avg INcome")
plt.show()
```
